Задача 44: В ячейке ниже представлен код генерирующий DataFrame, которая состоит всего
из 1 столбца. Ваша задача перевести его в one hot вид. Сможете ли вы это сделать без
get_dummies?

```python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data
```

Результат применения get_dummies:

```python
pd.get_dummies(data['whoAmI'])
```

Тот же результат без get_dummies:

```python
data['human']=list(map(lambda x: 1 if x=='human' else 0,lst))
data['robot']=list(map(lambda x: 1 if x=='robot' else 0,lst))
```

```python
data.drop(columns=['whoAmI'])
```
